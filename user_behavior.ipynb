```python
import pandas as pd
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pickle

df = pd.read_csv('/content/drive/My Drive/dataset/user_behavior.csv')
df.head(5)
```

```python
label_encoder = LabelEncoder()
```

```python
# Encoding untuk kolom 'Device Model', 'Operating System', dan 'Gender'
df['Device Model'] = label_encoder.fit_transform(df['Device Model'])
df['Operating System'] = label_encoder.fit_transform(df['Operating System'])
df['Gender'] = label_encoder.fit_transform(df['Gender'])
```

```python
print(df.head(5))
```

```python
# Pilih fitur (independent variables) dan label (dependent variable)
X = df.iloc[:, 1:-1]  # Semua kolom kecuali kolom User ID dan kolom terakhir sebagai fitur
y = df.iloc[:, -1]    # Kolom terakhir sebagai label
```

```python
# Membagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
X_train.head()
```

```python
# Membuat model DecisionTreeClassifier
model = tree.DecisionTreeClassifier()

# Melatih model
model = model.fit(X_train, y_train)

# Prediksi menggunakan data uji
y_pred = model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score: ", accuracy)
```

```python
# Prediksi untuk data baru
data_baru = [[3, 0, 393, 6.4, 1872, 67, 1122, 40, 1]]  # Google Pixel 5, Android, Male
data_baru2 = [[1, 0, 268, 4.7, 1331, 42, 944, 47, 0]]  # OnePlus 9, Android, Female

hasil = model.predict(data_baru)[0]
print("Hasil Prediksi: ", hasil)
```
